```python
from helper import load_training_log, read_epoch

log_ft = load_training_log("training_log/training_log_rubber_batch.txt")
log_fz = load_training_log("training_log/training_log_rubber_frozen_batch.txt")
```

```python
epoch = 200
print(read_epoch(log_ft, "Validation", epoch))
print(read_epoch(log_fz, "Validation", epoch))
```

```python
log_data = log_ft

import matplotlib.pyplot as plt

# Extract training and validation loss values
# train_losses = [log_data[epoch]['Training']['train_accuracy_class_8'] for epoch in log_data if 'Training' in log_data[epoch]][2:]
validation_losses = [log_data[epoch]['Validation']['validation_accuracy_class_7'] for epoch in log_data if 'Validation' in log_data[epoch]][1:]

# Generate the epochs list (assuming continuous epochs)
epochs = range(1, 200, 5)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(epochs, validation_losses, label='Validataion Accuracy', color='blue', marker='o')
# plt.plot(epochs, validation_losses, label='Validation Loss', color='red', marker='x')

# Adding title and labels
plt.title('Rubber Classification Accuracy (Validation)')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
# plt.ylim(0.8, 1.04)
plt.legend()

# Show the plot
plt.show()
```
